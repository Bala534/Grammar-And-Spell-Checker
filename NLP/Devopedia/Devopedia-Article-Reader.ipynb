```python
ArticleFile = '/content/drive/My Drive/NLP-Resources/Devopedia/devopediaArticles.v2525.json'
```

```python
import pandas as pd
```

```python
articles = pd.read_json(ArticleFile, orient='index')
```

```python
# Expand secs column to get References column
secs = articles['secs'].apply(pd.Series)
articles = pd.concat([articles, secs], axis=1)
articles = articles[~articles.References.isna()]
```

```python
articles.head()
```

```python
def convert_article_refs(refs):
    return tuple(ref['url'] if 'url' in ref else None for ref in refs)
```

```python
articles['urls'] = articles['References'].apply(convert_article_refs)
```

```python
articles.columns
```

```python
# Export all the reference URLs if required for other tasks
import json
urls = set(url for urls in articles.loc[:,'urls'].ravel() for url in urls if url)
with open('urls.json', 'w') as outfile:
    json.dump(list(urls), outfile, indent=2)
```
